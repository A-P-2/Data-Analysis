```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn import svm
from sklearn.datasets import make_blobs
from sklearn.covariance import EllipticEnvelope
from sklearn.ensemble import IsolationForest
```

# Составление набора данных

```python
# data 1
X1, y_true1 = make_blobs(n_samples=50, centers=1,
                       cluster_std=0.5, random_state=5)

X1[0] = [-10, 5]
X1[1] = [-4, 6]
X1[2] = [-6, 10]

#data 2
X2, y_true2 = make_blobs(n_samples=100, centers=2,
                       cluster_std=0.65, random_state=3)

X2[0] = [-5, 5]
X2[3] = [-2.2, 4]
X2[2] = [1.2, 2.5]

figure, axis = plt.subplots(1, 2, figsize=(15, 5))

axis[0].scatter(X1[:, 0], X1[:, 1], c=y_true1, s=50, cmap='viridis');
axis[1].scatter(X2[:, 0], X2[:, 1], c=y_true2, s=50, cmap='viridis');
```

# Выявление аномалий

**Метод Robust covariance**

```python
rc_pred1 = EllipticEnvelope(contamination=0.06).fit_predict(X1)
rc_pred2 = EllipticEnvelope(contamination=0.06).fit_predict(X2)

figure, axis = plt.subplots(1, 2, figsize=(15, 5))

axis[0].scatter(X1[:, 0], X1[:, 1], c=rc_pred1, s=50, cmap='bwr');
axis[1].scatter(X2[:, 0], X2[:, 1], c=rc_pred2, s=50, cmap='bwr');
```

**Метод One-Class SVM**

```python
oc_pred1 = svm.OneClassSVM(nu=0.06).fit_predict(X1)
oc_pred2 = svm.OneClassSVM(nu=0.06).fit_predict(X2)

figure, axis = plt.subplots(1, 2, figsize=(15, 5))

axis[0].scatter(X1[:, 0], X1[:, 1], c=oc_pred1, s=50, cmap='bwr');
axis[1].scatter(X2[:, 0], X2[:, 1], c=oc_pred2, s=50, cmap='bwr');
```

**Метод Isolation Forest**

```python
if_pred1 = IsolationForest(contamination=0.06).fit_predict(X1)
if_pred2 = IsolationForest(contamination=0.06).fit_predict(X2)

figure, axis = plt.subplots(1, 2, figsize=(15, 5))

axis[0].scatter(X1[:, 0], X1[:, 1], c=if_pred1, s=50, cmap='bwr');
axis[1].scatter(X2[:, 0], X2[:, 1], c=if_pred2, s=50, cmap='bwr');
```
