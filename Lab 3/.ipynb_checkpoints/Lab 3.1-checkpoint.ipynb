```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans, DBSCAN
from sklearn_extra.cluster import KMedoids
from sklearn.metrics import silhouette_score
from sklearn.model_selection import train_test_split
```

```python
X, y_true = make_blobs(n_samples=300, centers=5,
                       cluster_std=0.85, random_state=9)
plt.scatter(X[:, 0], X[:, 1], c=y_true, s=50, cmap='viridis');
```

```python
X2_1 = np.array([[np.sin(np.radians(i)) + np.random.uniform(-0.25, 0.25),
                np.cos(np.radians(i)) + np.random.uniform(-0.25, 0.25)] 
               for i in range(0, 360 * 5, 10)])

X2_2 = np.array([[np.sin(np.radians(i)) * 2 + np.random.uniform(-0.25, 0.25),
                np.cos(np.radians(i)) * 2 + np.random.uniform(-0.25, 0.25)] 
               for i in range(0, 360 * 10, 10)])

X2 = np.concatenate((X2_1, X2_2))
plt.scatter(X2[:, 0], X2[:, 1], s=50);
```

#  

**  **

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 12))
sub = 0

for i in range(3, 10, 2):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(X)
    y_kmeans = kmeans.predict(X)
    
    axs[sub // 2, sub % 2].scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
    axs[sub // 2, sub % 2].set_title(str(i) + '-means')
    
    sub += 1
```

```python
def GetNoisedData(ratio):
    X_normal, X_noised, y_normal, y_noised = train_test_split(X, y_true, test_size=ratio, random_state=1)
    
    for i in range(len(X_noised)):
        X_noised[i][0] += np.random.uniform(-2, 2)
        X_noised[i][1] += np.random.uniform(-2, 2)
        
    return np.concatenate((X_normal, X_noised)), np.concatenate((y_normal, y_noised))
```

**  (20%)**

```python
new_X, new_y = GetNoisedData(0.2)

plt.scatter(new_X[:, 0], new_X[:, 1], c=new_y, s=50, cmap='viridis');
```

**  **

```python
fig, axs = plt.subplots(4, 2, figsize=(15, 24))
sub = 0

for i in range(3, 10, 2):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(new_X)
    y_kmeans = kmeans.predict(new_X)
    
    kmedoids = KMedoids(n_clusters=i)
    kmedoids.fit(new_X)
    y_kmedoids = kmedoids.predict(new_X)
    
    axs[sub, 0].scatter(new_X[:, 0], new_X[:, 1], c=y_kmeans, s=50, cmap='viridis')
    axs[sub, 0].set_title(str(i) + '-means')
    
    axs[sub, 1].scatter(new_X[:, 0], new_X[:, 1], c=y_kmedoids, s=50, cmap='viridis')
    axs[sub, 1].set_title(str(i) + '-medoids')
    
    sub += 1
```

**  **

```python
fig, axs = plt.subplots(4, 2, figsize=(15, 24))
sub = 0

for i in range(3, 10, 2):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(X2)
    y_kmeans = kmeans.predict(X2)
    
    kmedoids = KMedoids(n_clusters=i)
    kmedoids.fit(X2)
    y_kmedoids = kmedoids.predict(X2)
    
    axs[sub, 0].scatter(X2[:, 0], X2[:, 1], c=y_kmeans, s=50, cmap='viridis')
    axs[sub, 0].set_title(str(i) + '-means')
    
    axs[sub, 1].scatter(X2[:, 0], X2[:, 1], c=y_kmedoids, s=50, cmap='viridis')
    axs[sub, 1].set_title(str(i) + '-medoids')
    
    sub += 1
```

```python
fig, axs = plt.subplots(4, 3, figsize=(17.5, 18))
sub = 0

for i in range(3, 10, 2):
    eps = 0.1
    dbscan = DBSCAN(min_samples=i, eps=eps)
    y_dbscan = dbscan.fit_predict(X2)
    
    axs[sub, 0].scatter(X2[:, 0], X2[:, 1], c=y_dbscan, s=50, cmap='viridis')
    axs[sub, 0].set_title('min_samples: %d, eps: %.1f' % (i, eps))
    
    eps = 0.25
    dbscan = DBSCAN(min_samples=i, eps=eps)
    y_dbscan = dbscan.fit_predict(X2)
    
    axs[sub, 1].scatter(X2[:, 0], X2[:, 1], c=y_dbscan, s=50, cmap='viridis')
    axs[sub, 1].set_title('min_samples: %d, eps: %.1f' % (i, eps))
    
    eps = 0.5
    dbscan = DBSCAN(min_samples=i, eps=eps)
    y_dbscan = dbscan.fit_predict(X2)
    
    axs[sub, 2].scatter(X2[:, 0], X2[:, 1], c=y_dbscan, s=50, cmap='viridis')
    axs[sub, 2].set_title('min_samples: %d, eps: %.1f' % (i, eps))
    
    sub += 1
```
